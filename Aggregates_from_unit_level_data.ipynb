```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#(32, 34), (34, 36),
colspecs= [(0, 32),  (36,39), (40,42),(42,45), (45,50),(50,55),(55,56), (56,57), (57,58), (58,61),(61,62),(62,64)  ]

fwf=  pd.read_fwf("/content/drive/MyDrive/Dataset/Time Use survey /Copy of TUS106_L05.txt",colspecs=colspecs, low_memmory=True, header=None)
```

```python
fwf.columns=["Common_ID","Serial_no_of_member","age", "srl__No_of_activity","time_from", "time_to", 
             "whether_performed_multiple_activ","whether_simultaneous_activity", "whether_a_major_activity",
            "_3_didit_activity_code",  "where_the_activity_was_performed", "unpaid_paid_status_of_activity" ]
```

```python
fwf=fwf[(fwf._3_didit_activity_code>300) & (fwf._3_didit_activity_code<499)]
```

```python
level2 = pd.read_stata("/content/drive/MyDrive/Dataset/Time Use survey /stata/Copy of TUS106_L02.dta")
level2["state"] = [i[0:2] for i in level2["NSS_Region"]]
level2["Person_serial_no_"]=level2["Person_serial_no_"].astype("float")

level2
```

```python
individual_data = level2[["Common_ID","Person_serial_no_" ,"state","Sector", "Gender","usual_principal_activity__status", "industry_of_work__2_digit_of_NIC", "weight_level2"]]
```

```python

fwf = fwf.merge(individual_data, right_on=["Common_ID","Person_serial_no_"],
          left_on=["Common_ID","Serial_no_of_member" ], how="left")
```

```python

fwf["time_spent"] = pd.to_datetime(fwf.time_to, format="%H:%M") - pd.to_datetime(fwf.time_from, format="%H:%M")

fwf["time_spent"]= fwf["time_spent"].dt.seconds/60
```

```python
aggregate = fwf.groupby(["state","Sector","Gender", "Common_ID","Serial_no_of_member"])["time_spent", "weight_level2"].agg({"time_spent":"sum","weight_level2":"mean" })
```

```python
major_fwf = fwf[(fwf.whether_a_major_activity==1) & (fwf.age.between(15,60) ) ] # 
```

```python

aggregate = major_fwf[major_fwf.weight_level2!=0].groupby(["state","Gender", "Common_ID","Serial_no_of_member"])[["time_spent"]].agg("sum")


aggregate=aggregate.reset_index()
aggregate = aggregate.merge(fwf[["Common_ID", "weight_level2"]])

aggregate = aggregate.groupby(["state", "Gender"])[["time_spent", "weight_level2"]].apply(lambda gp: np.average(gp.time_spent, weights=gp.weight_level2))

time_spent_sectorwise= aggregate/(60*24)
time_spent_sectorwise=time_spent_sectorwise.reset_index()

time_spent = time_spent_sectorwise
```

```python
#Time spent for wholle of India

aggregate = major_fwf[major_fwf.weight_level2!=0].groupby(["Gender", "Common_ID","Serial_no_of_member"])[["time_spent"]].agg("sum")


aggregate=aggregate.reset_index()
aggregate = aggregate.merge(fwf[["Common_ID", "weight_level2"]])

aggregate = aggregate.groupby(["Gender"])[["time_spent", "weight_level2"]].apply(lambda gp: np.average(gp.time_spent, weights=gp.weight_level2))

time_spent_india= aggregate/(60*24)
time_spent_india=time_spent_india.reset_index()

time_spent_india= time_spent_india.reset_index()

time_spent_india
```
